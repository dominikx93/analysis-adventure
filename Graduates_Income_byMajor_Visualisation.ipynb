Import libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Read a CSV file into DataFrame;
<br>
Examine DataFrame & its first 5 rows.

```python
df1 = pd.read_csv('degrees-that-pay-back.csv')
df1.head()
```

Rewrite columns' name with list of names;
<br>
Examine DataFrame & its first 5 rows.

```python
df1.columns = ['major', 'bgn_Q2', 'mid_Q2', 'change_bgn_mid', 'mid_p10', 'mid_Q1', 'mid_Q3', 'mid_p90']
df1.head()
```

Remove unnecessary dollar signs and commas from specified columns;
<br>
Convert the resulting strings to numeric values;
<br>
Examine DataFrame & its first 5 rows.

```python
dollar_cols = ['bgn_Q2', 'mid_Q2', 'mid_p10', 'mid_Q1', 'mid_Q3', 'mid_p90']

for x in dollar_cols:
    df1[x] = df1[x].str.replace('$','')
    df1[x] = df1[x].str.replace(',','')
    df1[x] = pd.to_numeric(df1[x])

df1.head()
```

Generate descriptive statistics from the DataFrame.

```python
df1.describe()
```

Sort the DataFrame in place by bgn_Q2 column in descending order;
<br>
Reset its index;
<br>
Examine DataFrame & its first 5 rows.

```python
df1.sort_values(by = 'bgn_Q2', ascending = False, inplace = True)
df1 = df1.reset_index()
df1.head()
```

Create a figure and axes with specified size;
<br>
Set axes background colour;
<br>
Assign values from coloumn bgn_Q2 to x_bgn;
<br>
Compute y value based on the index;
<br>
Set labels from major column;
<br>
Scatter plot x_bgn vs y with red downward triangles with specified size, label as Median Starting Salary;
<br>
Set y-tick labels from labels;
<br>
Assign values from coloumn mid_Q2 to x_mid;
<br>
Scatter plot x_mid vs y with yellow diamonds with specified size, label as Median Mid-Career Salary;
<br>
Label the x axis as USD $, set title, legend, grid, and display the plot.

```python
fig, ax = plt.subplots(figsize=(8, 12))
ax.set_facecolor('#c6c6c6')

x_bgn = df1['bgn_Q2']
y = len(df1.index) - df1.index
labels = df1['major']

plt.scatter(x_bgn, y, s = 50, color = '#e64626', marker = 'v', label = 'Median Starting Salary')
plt.yticks(y, labels)

x_mid = df1['mid_Q2']

plt.scatter(x_mid, y, s = 60, color = '#ffe600', marker = 'D', label = 'Median Mid-Career Salary')

plt.xlabel('USD $')
plt.ylabel('')
plt.title('Salary Information by Major (Sorted by Median Starting Salary)')
plt.legend(loc = 2)
plt.grid(True)
plt.show()
```

Almost identical to the previous one;
<br>
Except that it sorts the DataFrame in descending order by the mid_Q2 column, resets the index, thereby changing the order of majors on the y axis.

```python
df2 = df1.sort_values(by = 'mid_Q2', ascending = False)
df2 = df2.reset_index()

fig, ax = plt.subplots(figsize=(8, 12))
ax.set_facecolor('#c6c6c6')

x_bgn = df2['bgn_Q2']
y = len(df2.index) - df2.index
labels = df2['major']

plt.scatter(x_bgn, y, s = 50, color = '#e64626', marker = 'v', label = 'Median Starting Salary')
plt.yticks(y, labels)

x_mid = df2['mid_Q2']
plt.scatter(x_mid, y, s = 60, color = '#ffe600', marker = 'D', label = 'Median Mid-Career Salary')

plt.xlabel('USD $')
plt.ylabel('')
plt.title('Salary Information by Major (Sorted by Median Mid-Career Salary)')
plt.legend(loc = 2)
plt.grid(True)
plt.show()
```

It is still similar with previous ones;
<br>
Except multiple x values (mid_Q1, mid_Q1 and mid_Q3 columns) are plotted vs y here and overlapping trendlines are computed and displayed for all three.

```python
fig, ax = plt.subplots(figsize=(8, 12))
ax.set_facecolor('#c6c6c6')

y = len(df2.index) - df2.index
labels = df2['major']
plt.yticks(y, labels)

x_mid_Q1 = df2['mid_Q1']
plt.scatter(x_mid_Q1, y, s = 60, color = '#0f4beb', marker = 'v', label = 'Q1 Mid-Career Salary')
trend_Q1 = np.poly1d(np.polyfit(y, x_mid_Q1, 1))
plt.plot(trend_Q1(y), y, color = '#0f4beb', linestyle = '-', label = 'Trendline (Q1)')

x_mid_Q2 = df2['mid_Q2']
plt.scatter(x_mid_Q2, y, s = 50, color = '#e64626', marker = 'D', label = 'Q2 Mid-Career Salary')
trend_Q2 = np.poly1d(np.polyfit(y, x_mid_Q2, 1))
plt.plot(trend_Q2(y), y, color = '#e64626', linestyle = '-', label = 'Trendline (Q2)')

x_mid_Q3 = df2['mid_Q3']
plt.scatter(x_mid_Q3, y, s = 60, color = '#ffe600', marker = '^', label = 'Q3 Mid-Career Salary')
trend_Q3 = np.poly1d(np.polyfit(y, x_mid_Q3, 1))
plt.plot(trend_Q3(y), y, color = '#ffe600', linestyle = '-', label = 'Trendline (Q3)')

plt.xlabel('USD $')
plt.ylabel('')
plt.title('Salary Information by Major (Quartiles)')
plt.legend(loc=2)
plt.grid(True)
plt.show()
```
